```python
import tensorflow as tf
import numpy as np 
import time , os
```

```python
text = open('./khayyammolana/khayyam.txt', 'rb').read().decode(encoding='utf-8')
```

```python
vocab = sorted(set(text))
```

```python
vocab2index = {i:u for u , i  in enumerate(vocab)}
index2vocan = np.array(vocab)
vocab2index
```

```python
text_as_integer = np.array([ vocab2index[i] for i in text ])
```

```python
datasets = tf.data.Dataset.from_tensor_slices(text_as_integer)
```

```python
sequences = datasets.batch(30 , drop_remainder=True )
sequences
```

```python
def make_sequence_for_GRU(batch):
    sequence = batch[:-1]
    target = batch[1:]
    return sequence , target
```

```python
sequences  = sequences.map(make_sequence_for_GRU)
```

```python
sequences
```

```python
dataset = sequences.batch(64 , drop_remainder=True )
```

```python
dataset
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(vocab), 25),
    tf.keras.layers.GRU(1024, return_sequences=True),
    tf.keras.layers.Dense(len(vocab))
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam' , loss='sparse_categorical_crossentropy')
```

```python
chechpoint = tf.keras.callbacks.ModelCheckpoint('khayyam/checkpoints',   save_weights_only=True)
```

```python
model.fit(dataset , epochs=5 , callbacks=[chechpoint])
```
